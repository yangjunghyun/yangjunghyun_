```python
#!pip install kmeans_pytorch
import pandas as pd
from sklearn.model_selection import train_test_split
import torch
from kmeans_pytorch import kmeans, kmeans_predict
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/KOPIS/pre_processing_data/2019_07_09/df2019_07_09.csv')
df.info()
print('-----------------------------------')
print(df)
```

```python
# 결측값 처리
data= pd.get_dummies(df, columns=['장르명'], dummy_na=True)
print(data)
```

```python
from sklearn.model_selection import train_test_split
x, y = train_test_split(data, test_size=0.2, random_state=123)
```

```python
if torch.cuda.is_available():
    device = torch.device('cuda:0')
else:
    device = torch.device('cpu')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit(data).transform(x)
y_scaled = scaler.fit(y).transform(y)
```

```python
x = torch.from_numpy(X_scaled)
y = torch.from_numpy(y_scaled)
```

```python
print(x.size())
print(y.size())
print(x)
```

```python
num_clusters = ? # 우리 변수 몇개일까 과연
cluster_ids_x, cluster_centers = kmeans(
    X=x, num_clusters=num_clusters, distance='euclidean', device=device
)
```

```python
print(cluster_ids_x)
print(cluster_centers)
```

```python
cluster_ids_y = kmeans_predict(
    y, cluster_centers, 'euclidean', device=device
)
```

```python
print(cluster_ids_y)
```

```python
plt.figure(figsize=(4, 3), dpi=160)
plt.scatter(y[:, 0], y[:, 1],c=cluster_ids_y, cmap='viridis', marker='x')

plt.scatter(
    cluster_centers[:, 0], cluster_centers[:, 1],
    c='white',
    alpha=0.6,
    edgecolors='black',
    linewidths=2
)

plt.tight_layout()
plt.show()
```
