```python
def groupby_age(x):
    if x <= 13:
        score = '0~13세'
    elif x <= 19:
        score = '14~19세'
    elif x < 30:
        score = '20대'
    elif x < 40:
        score = '30대'
    elif x < 50:
        score = '40대'
    elif x < 60:
        score = '50대'
    elif x < 70:
        score = '60대'                 
    else:
        score = '70세 이상'
    return score
```

```python
def pre_processing(exdata):
  exdata = exdata[exdata['연령']!=0]
  exdata = exdata[exdata['성별']!=0]
  exdata = exdata[exdata['연령']<=2021] # 이상치 제거
  del exdata['출연진내용']
  del exdata['제작진내용']
  del exdata['오픈런 여부']
  return exdata
```

```python
def age(exdata):
  exdata['나이'] = 2022 - exdata['연령'] + 1 # 나이 계산
  exdata["연령대"] = exdata["나이"].apply(lambda x: groupby_age(x))
  return exdata
```
