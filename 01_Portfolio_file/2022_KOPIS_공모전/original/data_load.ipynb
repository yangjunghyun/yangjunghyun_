# 2번 주제

```python
import pandas as pd

# 데이터 불러오기
data_02 = pd.read_csv('/content/drive/MyDrive/KOPIS/pre_processing_data/2019_07_09/df2019_07_09_2.csv')
data_02.head()
```

```python
### 데이터 전처리 함수
def pre_processing(exdata):
  # 결측치 제거
  exdata = exdata.dropna()
  # vip석만 추출
  exdata = exdata[exdata['좌석등급'].str.contains('VIP석')]
  # 장당금액 10만원 이상만 추출
  exdata = exdata[exdata['장당금액']>=100000]
  # 장당금액 30만원 이하만 추출
  exdata = exdata[exdata['장당금액']<=300000]
  # 예매/취소방식명(관리시스템)이 모바일과 웹인 경우만 추출
  exdata = exdata[exdata['예매/취소방식코드'].isin([1, 2])]
  # 시간데이터 datetime 형식 변경해주기
  exdata['예매/취소일시'] = pd.to_datetime(exdata['예매/취소일시'])
  # 바꾼 데이터로 날짜 분리하기
  exdata['날짜'] = exdata['예매/취소일시'].dt.strftime('%Y%m%d')
  exdata['월'] = exdata['예매/취소일시'].dt.strftime('%m')
  exdata['일'] = exdata['예매/취소일시'].dt.strftime('%d')
  return exdata
```

```python
# 함수적용시키기
data_19_07_09 = pre_processing(data_02)
```
