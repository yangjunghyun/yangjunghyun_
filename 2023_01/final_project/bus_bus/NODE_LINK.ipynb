```python
from shapely.geometry import Point
from shapely.geometry import LineString
import geopandas as gpd

# 좌표 변경 함수 정의
def swap_coordinates(geom):
    lon, lat = geom.x, geom.y
    return Point(lat, lon)  # 위도와 경도 순서를 바꾼다.

# GeoDataFrame 생성
gdf = gpd.GeoDataFrame(BUS_SPEED_DATA_30300037_MERGE, geometry='START_NODE_GEOMETRY')

# 좌표 변경 적용
gdf['START_NODE_GEOMETRY'] = gdf['START_NODE_GEOMETRY'].apply(swap_coordinates)
gdf['END_NODE_GEOMETRY'] = gdf['END_NODE_GEOMETRY'].apply(swap_coordinates)

gdf['line'] = gdf.apply(lambda row: LineString([row['START_NODE_GEOMETRY'], row['END_NODE_GEOMETRY']]), axis=1) #Create a linestring column
```
