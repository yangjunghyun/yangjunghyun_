```python
result = []
for row in data[2:]:
    result.append(sum(row[10:15:2]))
len(result)
```

```python
# import matplotlib.pyplot as plt
result.sort()
plt.style.use('ggplot')
plt.figure(figsize=(10,5))
plt.bar(range(len(result)),result)
plt.show()
```

```python
mx = 0
mx_station = ''

for row in data[2:]:
    if sum(row[10:15:2]) > mx:
        mx = sum(row[10:15:2])
        mx_station = row[3]+ ' '+ row[1]
print(mx_station, mx)
```

```python
mx = 0
mx_station = ''
t = int(input("몇 시의 승차 인원이 가장 많은 역이 궁금하세요?"))
for row in data[2:]:
    a = row[4+(t-4)*2]
    #a = row[5+(t-4)*2] 하차인원
    if a > mx:
        mx = a
        mx_station = row[3]+ ' '+ row[1]
print(mx_station, mx)
```

```python
mx = [0]*24
mx_station = ['']*24
for row in data[2:]:
    for j in range(24):
        a = row[j*2+4]
        if a > mx[j]:
            mx[j] = a
            mx_station[j] = row[3]+' '+row[1]
for t, name in enumerate(mx_station):
    print(name, t+4, '시',mx[t])
```

```python
plt.rc('font',family = 'Malgun Gothic')
plt.bar(range(24),mx)
plt.xticks(range(24),mx_station,rotation=90)
plt.show()
```

```python
# 하차
mx = [0]*24
mx_station = ['']*24
for row in data[2:]:
    for j in range(24):
        a = row[j*2+5]
        if a > mx[j]:
            mx[j] = a
            mx_station[j] = row[3]+' '+row[1]
for t, name in enumerate(mx_station):
    print(name, t+4, '시',mx[t])
```

```python
plt.rc('font',family = 'Malgun Gothic')
plt.bar(range(24),mx)
plt.xticks(range(24),mx_station,rotation=90)
plt.show()
```

```python
# 승하차인원
s_in = [0]*24
s_out = [0]*24
for row in data[2:]:
    for j in range(24):
        s_in[j] += row[j*2+4]
        s_out[j] += row[j*2+5]
        
for t, name in enumerate(mx_station):
    print(name, t+4, '시',mx[t])
```

```python
plt.rc('font',family = 'Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label = '승차')
plt.plot(s_out, label = '하차')
plt.legend()
# plt.xticks(range(24),range(4:28))
plt.show()
```

```python
xpath = '//*[@id="SIGUNGU_NM0"]'
gu_list_raw = driver.find_element_by_xpath(xpath)
gu_list = gu_list_raw.find_elements_by_tag_name('option')
```

```python
gu_names = [option.get_attribute("value") for option in gu_list]
gu_names.remove('')
gu_names
```

```python
xpath = '//*[@id="searRgSelect"]'
element_sel_gu = driver.find_element_by_xpath(xpath).click()
```

```python
xpath = '//*[@id="glopopd_excel"]'
element_get_excel = driver.find_element_by_xpath(xpath).click()
```

```python
import time
from tqdm import tqdm_notebook

for gu in tqdm_notebook(gu_names):
    element = driver.find_element_by_id("SIGUNGU_NM0")
    element.send_keys(gu)
    
    time.sleep(2)
    
    xpath = '//*[@id="searRgSelect"]'
    element_sel_gu = driver.find_element_by_xpath(xpath).click()
    
    time.sleep(1)
    
    xpath = '//*[@id="glopopd_excel"]'
    element_get_excel = driver.find_element_by_xpath(xpath).click()
    
    time.sleep(1)
```
