```python
bd_temp = []
for row in result:
    month = row[0].split('/')[1]
    day = row[0].split('/')[2]
    if month == '01' and day == '20':
        bd_temp.append(row[2])
len(bd_temp)
```

```python
head = ['일시','지점번호','평균기온(℃)', '최저기온(℃)', '최고기온(℃)']
bd_temp_avg = []
bd_temp_min = []
bd_temp_max = []
yearl = []
for row in result:
    year = row[0].split('/')[0]
    month = row[0].split('/')[1]
    day = row[0].split('/')[2]
    if month == '01' and day == '20':
        yearl.append(year)
        bd_temp_avg.append(row[2])
        bd_temp_min.append(row[3])
        bd_temp_max.append(row[4])
len(bd_temp)
```

```python
plt.plot(bd_temp_avg)
```

```python
plt.rc('font',family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
# plt.style.use('ggplot')
# plt.figure(figsize=(10,2),dpi=300)
plt.title('내 생일날 온도 변화')
plt.plot(yearl,bd_temp, label = '평균기온')
plt.plot(yearl,bd_temp_min, label = '최저기온')
plt.plot(yearl,bd_temp_max, label = '최고기온')
plt.xlabel('년도')
plt.ylabel('온도')
# plt.savefig('내생일온도변화.png')
plt.legend()
plt.show()
```

### 제주도의 성별 인구 비율 표현하기

```python
m = 0
f = 0
name = input('찾고 싶은 지역의 이름을 입력하세요 : ')
for row in g[2:]:
    if name in row[0]:
        for i in range(0,101):
            m = m + (row[i+3])
            f = f + (row[i+106])
        break
plt.figure(figsize = (10,5))
size = [m,f]
label = ['남자','여자']
color = ['crimson','darkcyan']
plt.pie(size,labels = label, colors = color, autopct = '%.1f%%', startangle=90)
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.legend()
plt.show()
```

### 꺾은선 그래프

```python
import usecsv
result = usecsv.switch(usecsv.opencsv('gender.csv'))
m = []
f = []
name = input('궁금한 동네를 입력하세요 : ')
for row in result[2:]:
    if name in row[0]:
        for i in range(0,101):
            m.append(row[i+3])
            f.append(row[i+106])
        break
plt.plot(m, label = 'Male')
plt.plot(f, label = 'Female')
plt.legend()
plt.show()
```

### 막대그래프

```python
m = []
f = []
age = []
name = input('궁금한 동네를 입력하세요 : ')
for row in result[2:]:
    if name in row[0]:
        for i in range(0,101):
            m.append(row[i+3])
            f.append(row[i+106])
            age.append(int(row[i+3])-int(row[i+106]))
        break
plt.bar(range(101),age)
plt.show()
```
