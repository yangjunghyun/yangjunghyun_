### numpy 함수

```python
a = np.arange(-np.pi,np.pi,np.pi/100)
plt.plot(a, np.sin(a))
plt.show()
```

```python
a = np.arange(-np.pi,np.pi,np.pi/100)
plt.plot(a, np.sin(a), label = 'np.sin')
plt.plot(a, np.cos(a), label = 'np.sin')
plt.plot(a+np.pi/2, np.sin(a), label = 'np.sin+1.7')
plt.legend()
plt.show()
```

### 조건연산 (mask)

```python
a = np.arange(-5,5)
a
```

```python
a < 0
```

```python
a[a<0]
```

```python
# abs 절대값
mask1 = abs(a) > 3
a[mask1]
```

```python
mask2 = abs(a) % 2 == 0
a[mask2]
```

```python
# or 연산
mask1 + mask2
```

```python
# and 연산
mask1 * mask2
```

```python
a[mask1+mask2]
```

```python
x = np.random.randint(-100,100,1000)
y = np.random.randint(-100,100,1000)
mask1 = abs(x) > 50
mask2 = abs(y) > 50
x = x[mask1+mask2] # or연산
y = y[mask1+mask2]
size = np.random.rand(len(x))*100
plt.scatter(x,y,s =size, c= x, cmap = 'jet',alpha = 0.7)
plt.colorbar()
plt.show()
```
