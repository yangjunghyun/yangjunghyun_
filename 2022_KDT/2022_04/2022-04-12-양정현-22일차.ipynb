````python
news = 'https://news.daum.net'
soup = bs(ur.urlopen(news).read(), 'html.parser')
headline = soup.find_all('div', {'class':'item_issue'})
for i in headline:
    #print(i.text.replace("\n",'')[38:-32],  '\n')
    print(i.text.replace("\n",'').replace('  ','')[2:],  '\n')
    url = i.find_all('a')[0].get('href')
    soup3 = bs(ur.urlopen(url).read(), 'html.parser')
    for i in soup3.find_all('p'):#[10:-4]:
        if re.match('[가-힣]+',i.text):
            print(i.text)
    #print()
````

````python
import os, codecs, re, datetime, requests
import urllib.request as ur
from bs4 import BeautifulSoup as bs

url = 'https://news.daum.net'
soup = bs(ur.urlopen(url).read(),'html.parser')
f = open(str(datetime.date.today())+ 'article_total.txt','w')
for i in soup.find_all('div', {'class':'item_issue'}):
    try:
        f.write(i.text+'\n') #제목
        f.write(i.find_all('a')[0].get('href')+'\n') # 기사 URL
        soup2 = bs(ur.urlopen(i.find_all('a')[0].get('href').read(),'html.parser'))
        for j in soup2.find_all('p'):
            if re.match('[가-힣]+',j.text):
                f.write(j.text)
    except:
        pass
f.close()
````

````python
import os, codecs, re, datetime, requests
import urllib.request as ur
from bs4 import BeautifulSoup as bs

url = 'https://news.daum.net/'
soup = bs(ur.urlopen(url).read(), 'html.parser')
f = open(str(datetime.date.today())+'article_total.txt', 'w')
for i in soup.find_all('div', {'class':'item_issue'}):
    try:
        f.write(i.text+'\n') #제목
        #f.write(i.text.replace("\n",'').replace('  ','')[2:],  '\n')
        f.write(i.find_all('a')[0].get('href')+'\n') #기사URL
        soup2 = bs(ur.urlopen(i.find_all('a')[0].get('href')).read(), 'html.parser')
        for j in soup2.find_all('p'):#[10:-4]:
            if re.match('[가-힣]+',j.text):
                f.write(j.text)
    except:
        pass

f.close()
````
